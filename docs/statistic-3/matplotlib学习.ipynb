```python
import pandas as pd
```

```python
df = pd.read_csv('./douban_movie.csv')
```

```python
df.head()
```

### 绘制每个国家或者地区的电影数量的柱状图

```python
# movies = df.groupby('产地').size().sort_values(ascending=False)
movies = df['产地'].value_counts()
```

```python
movies
```

```python
x = movies.index
y = movies.values

plt.figure(figsize=(15,6))

plt.bar(x, y, color='blue')
plt.title('每个国家或者地区的电影数量')

plt.xlabel('国家或地区', fontsize=20)
plt.ylabel('电影数量', fontsize=20)
plt.xticks(rotation=90, fontsize=15)

# plt.text?
for a,b in zip(x,y):
    plt.text(a, b+100, b, ha='center')

plt.show()
```

### 2. 绘制每年电影上映数量的曲线图

```python
df.head()
```

```python
res = df.groupby('年代').size()
```

```python
years_movie = res.sort_index()[:-2]
```

```python
years_movie
```

```python
import numpy as np
x = years_movie.index
y = years_movie.values

plt.figure(figsize=(10,6))
plt.xlabel('年份')
plt.ylabel('数量')
plt.xticks(np.arange(0,len(x),10))


for a,b in zip(x[::10],y[::10]):
    plt.text(a,b+50,b)

plt.plot(x, y)

plt.show()
```

### 3.根据电影的长度绘制饼图 

```python
df.head()
```

```python
# [0,30]  [30,90] [90,110], [110, 1000]
```

```python
pd.cut( np.array([0.2, 1.4, 2.5, 6.2, 9.7, 2.1]),[1,2,3] )  # (1,2]  (2,3]
```

```python
s = np.array(df['时长'])
'8U' in s
np.where('8U' == s)
```

```python
'12J' in s
np.where('12J' ==  s)
```

```python
s = np.delete(s, 31644, axis=0)
s = np.delete(s, 32947, axis=0)
np.where('12J' == s)
```

```python
data = pd.cut(s.astype('float'), [0,60,90,110,1000]).value_counts()
data
```

```python
x = data.index
y = data.values

plt.figure(figsize=(10,6))
plt.title('电影时长分布图')
patchs, l_text, p_text = plt.pie(y, labels=x, autopct='%0.2f%%', colors='bgry', startangle=90)

for i in p_text:
    i.set_size(15)
    i.set_color('w')

for l in l_text:
    l.set_size(20)
    l.set_color('r')
    
plt.show()
```
